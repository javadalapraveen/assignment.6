```python
class dog:

    def __init__(self,name=None,age=0,coat_color=None):
        self.name=name
        self.age=age
        self.coat_color=coat_color
        
    def description(self):
        self.name=input("enter the name: ")
        self.age=int(input("enter the age: "))
        print(f"name:{self.name},  age:{self.age}")

    def get_info(self):
        self.coat_color=input("enter the color: ")

        print(f"coat color: {self.coat_color}")
        

class jackrussellterrier(dog):
    def __init__(self, name=None, age=0, coat_color=None,energy_level=None):
        super().__init__(name, age, coat_color)
        self.energy_level=energy_level

    def bark(self):
        self.bark="ghrrr!, ghrrr!"
        print("barks like ",self.bark)

    def speed(self):
        self.energy_level=int(input("enter the level: "))
        self.speed=self.energy_level

        print("speed level is  ",self.speed)
        print("--------------------------------------------------->")

class Bulldog(dog):
    def __init__(self, name=None, age=0, coat_color=None,drool_level=0):
        super().__init__(name, age, coat_color)

        self.drool_level=drool_level

    def snore(self):

        self.snore="trrrrr..! trrrrr..!"

        print("snores like ",self.snore)

    def drool(self):
        self.drool="low"

        print("Drool ",self.drool)
        print("---------------------------------------------------->")

abbs=dog()
abbs.description()
abbs.get_info()
print("------------------------------------------------------->")
obbs=jackrussellterrier()
obbs.description()
obbs.get_info()
obbs.bark()
obbs.speed()
uff=Bulldog()
uff.description()
uff.get_info()
uff.snore()
uff.drool()
```
